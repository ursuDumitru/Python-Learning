# Linear Regression
( Supervised Learning )

One of the most basic machine learning models is the Linear Regression.
The purpose of linear regression is to take a bunch of data points and
to draw a straight line right through them. This line should then be as
near as possible to all the points. When we have found this line, we
can use it to make predictions for future values.
( https://www.neuralnine.com/linear-regression-from-scratch-in-python/ )

not 100% accurate
but generalised

We use this algorithm to make an ML model that predicts future values
( give a score and tell me the time it took to study or
  give a time and tell me the score it will most probably have )

```python
# TESTING

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def my_function(x, a, b):
    return a * x + b # for this functoin basically a tells the angle, and b just adds or subs from each y value (up or down)

x = np.linspace(0, 20)
y = my_function(x, 1, -20)

plt.plot(x, y, label='my_function', color='blue', linestyle='-', linewidth=2)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot my_function')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# reshaping the arrays into vertical for scikit-learn
time_studied = np.array([20, 50, 32, 65, 23, 43, 10, 5, 22, 35, 29, 5, 56]).reshape(-1,1)
scores = np.array([56, 83, 47, 93, 47, 82, 45, 23, 55, 67, 57, 4, 89]).reshape(-1, 1)

# print(time_studied)

model = LinearRegression()

model.fit(time_studied, scores) # our function y = a * x + b


hours_i_will_learn = 56
print(f'prediction of [{hours_i_will_learn}] hours, I get a score of : ' +
      f'{model.predict(np.array([hours_i_will_learn]).reshape(-1, 1))}')

plt.scatter(time_studied, scores)
plt.plot(np.linspace(0, 70, 100).reshape(-1, 1), model.predict(np.linspace(0, 70, 100).reshape(-1, 1)), 'r')
plt.xlabel('hours')
plt.ylabel('score')
plt.ylim(0, 100) # limit of y on plot
plt.show()
```

```python
# Testing data

from sklearn.model_selection import train_test_split

time_train, time_test, score_train, score_test = train_test_split(time_studied, scores, test_size=0.2) # 20% for testing

model = LinearRegression()
model.fit(time_train, score_train)

print(model.score(time_test, score_test))
print(model.predict(time_test))

plt.scatter(time_train, score_train)
plt.plot(np.linspace(0, 70, 100).reshape(-1, 1), # x
         model.predict(np.linspace(0, 70, 100).reshape(-1, 1)), # y
         'r')
```
