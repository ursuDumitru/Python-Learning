# Suport Vector Machines

Basically we have the same scenario as in K-Neighbors, but
    here we train a mathematical linear function to split the data



Infinite amount of ways to split the data, so we want the most
    generalized , optimal line (a line that is furthest away from
    all the points)

Uses support line which are parallel to the base line


Takes the nearest points and draws a line at equal distances ?


The space between the line and our points where there are no points
    is called 'Margin'


But data in real life isn't always grouped, so we will use a method
    called 'Kernnels'

This method uses functions that add a 3'rd dimension to our data
We hope thta somehow we can lift up the blue/red points to separate
    them form the others
We do not add new data, because kernels already have it
We are making the data more 'redundant'

Soft Margin ( tolerance )
not the best line:

soft-margin = 2 -> we allow up to 2 miss-clasifications
this being the best line:

By allowing some miss-classification you allow the model to be
    more precise in the long term
Also there should not be too many soft-margins


prettu cool:


```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC # Support Vector Machine
from sklearn.neighbors import KNeighborsClassifier

data = load_breast_cancer()

X = data.data
Y = data.target

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
# x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=23) # test/train remain the same

# kernel = rbf, polynomial which have diferent time/resources
clf = SVC(kernel='linear', C=3) # C = soft margin
clf.fit(x_train, y_train)
print(f'SVC : {clf.score(x_test, y_test)}')

clf2 = KNeighborsClassifier(n_neighbors=3)
clf2.fit(x_train, y_train)
print(f'KNeighbors : {clf2.score(x_test, y_test)}')
```
