# Decision Trees and Random Forest Classification

1. Decision Tree

We have a decision to predict using hystorical data
Example:
    Mark, goes to nature ? Yes : No
Let's say we have multiple hystorical events :

            Sunny   Windy   Rain    Alone   ...
    1. Yes  Yes     No      No      Yes
    2. No   Yes     Yes     No      No
    3. Yes  No      No      Yes     Yes
    4. Yes  ...     ...     ...     ...
    ...
    in which we know if Mark went to nature or not


If we have a decision tree which is already trained
    it would have a 'Root' and 2 paths


Each question/'Feature' may have multiple answers (branches)
    Example : How is the weather ? (Sunny, Rainy, Windy, ...)

We may have a data set :
        Sunny   Rainy   Alone   Conclusion
    1   Yes     ...             Yes
    2   Yes                     Yes
    3   No                      Yes
    4   No      ...             Yes
    ...
    If theoretically all Sunny ( Yes to Sunny ) have the conclusion ultimately
    'Yes' (ex: Yes: 85, No: 0) we could end the decision here not carring about
    the others, but if all not Sunny days ( No to Sunny ) have something like
    Yes: 33, No: 55 we continue asking questions


Also we have a random factor here because we do not know with which
    feature/question we should start
(the order of the features affects the output)


2. Random Forest Classification

Bassically these are multiple decision trees , trained on the same data
And ultimatelly we choose the colective result of those multiple trees
(multiple opinions on the same issue)

The advantage is minimising the risk of a miss-classification

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

data = load_breast_cancer()

X = data.data
Y = data.target

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

clf3 = DecisionTreeClassifier()
clf3.fit(x_train, y_train)
print(f'DecisionTreeClassifier : {clf3.score(x_test, y_test)}')

clf4 = RandomForestClassifier() # we could specify how many samples, how deep
clf4.fit(x_train, y_train)
print(f'RandomForestClassifier : {clf4.score(x_test, y_test)}')
```
