# K-Means Clustering

Unsupervised Learning Algorithm

We do not have labeled data(groups) to train on

(see the image all points are gray)

We are saying to the algorithm to find the groups for us (K-Means Clustering)
Here are the points, find the groups for us, and when we have a new point,
    assign it to one of the groups
Example of finding 2 clusters:


'Centroids' are the center of the clusters
Centroids are randomly chosen

So the algorithm looks at the data and assigns each point to the
    closest centroid
Example of grouping data to 3 clusters:


Then we delete the old centroids and we find the new centroids
    by taking the average of the points in each cluster(center of
    the points)

(see how the centroids moved)

Then we repeat the process until the centroids do not move anymore




```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale # the function used to normalize the data
from sklearn.datasets import load_digits # dataset of handwritten digits

digits = load_digits()
data = scale(digits.data) # normalize the data

# 10 (n_clusters) clusters because there are 10 digits
# 10 (n_init) iterations to find the best clusters
model = KMeans(n_clusters=10, init='random', n_init=10)
model.fit(data)

# we do not have any means to test the accuracy of the model
# because we do not have any labels for the data
# (unsupervised learning)

# the only measure of accuracy is how well the centroinds represent the data
# (the closer the centroids are to the data, the better the model)
model.predict([...])
```
