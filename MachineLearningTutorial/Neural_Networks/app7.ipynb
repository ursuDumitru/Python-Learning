# Neural Networks

In an Neural Network we have multiple layers of 'Neurons'
The first layer is the input layer, the last layer is the output layer

Input can be anything(data), but output is usually a classification or a regression
(reggresion is a continous output, classification is a discrete output)

How many neurons are on one layers depends on the problem we are trying to solve.
Meaning, how many features we have, and how many classes we have.

In between the input and output layers we have hidden layers.
The more hidden layers we have, the more complex the neural network is.

Each neuron of a layer is connected to every neuron of the next layer.
This means that each neuron of a layer has a weight for each neuron of the next layer.
The weights are the parameters of the neural network.
'Weights' is a matrix of shape (n_neurons_of_layer1, n_neurons_of_layer2)

The connections between the neurons are called 'synapses'.
And are randomly initialized at the beginning of the training.


If we have supervised learning, the training(feature) data is being
    tossed to the input layer, and then Training labels are being outputed

Example :
    We are going to classify handwritten digits from 0 to 9 and we expect
    8 as an answer ( digits means 10 classes -> 10 output neurons) and if
    the most activated neuron the 2nd one, then we say no, we have to adjust
    the weights of the synapses to get the correct answer.

    If a digit is a picture with the size 28 pixels by 28 pixels, it means
    that we need 784 input neurons (28*28=784).

    The output layer has 10 neurons, each neuron represents a digit from 0 to 9
    and each neuron has a value between 0 and 1, which represents the probability
    of the digit being the digit that the neuron represents.


Neurons:
    Every neuron has an input and an output.
    The input is the sum of the outputs of the previous layer multiplied by the weights
        of the corresponding synapses or just the input data if it is the first layer.
    The input gets feed into the neuron which triggers an activation function.
    The activation function is a function that decides whether the neuron should be
        activated or not(how bright the neuron is), based on whether each neuron's
        input is relevant for the model's prediction. (The activation function also
        transforms the input to get the output. There are different types of activation
        functions, and the choice of them affects the neural network's performance.)
    The function produces an activation value(for example 0.9), which gets sent out
        to the next layer, or in case of being the last layer, it is the output.
    


Weights:
    The weights essentially determine how much each input contributes to the neuron's
        output. Larger weights amplify the impact of the corresponding input, while
        smaller weights diminish it. This step captures the idea that some features
        might be more relevant than others in making predictions.
    If a weight is close to 0, then the input doesn't really matter for the next neuron
        and if it is close to 1, then the input is very important for the next neuron.
    
    (see w1 and w2 in the picture above)


Activation Functions:
1. Signoid function:
    The sigmoid function is a mathematical function used to map the predicted values
        to probabilities. It maps any real value into another value between 0 and 1.
        In machine learning, we use sigmoid to map predictions to probabilities.
    

2. ReLU function(Rectified Linear Unit):
    The ReLU function is another non-linear activation. The main advantage of using the
    ReLU function is that it does not activate all the neurons at the same time.
    If the input is negative, then the ReLU is zero, and if the input is positive,
    then the ReLU is equal to the input.
    


Gradient Descent:
    Gradient descent is an optimization algorithm used to minimize some function by
        iteratively moving in the direction of steepest descent as defined by the
        negative of the gradient. In machine learning, we use gradient descent to
        update the parameters of our model.


Loss Function:
    The loss function tells us how bad our model performs at each iteration.
        (how far the predicted value is from the actual value)
    Our goal is to minimize the loss function.
    We want to adjust the weights of the synapses in a way that the loss function
        is minimized.
    
    So we bassically take the gradient of the loss function and move in the opposite
        direction of the gradient(minimize the loss function) in 2 dimensions.
    
    In 3 dimensions we have to take the partial derivatives of the loss function
        with respect to each weight and move in the opposite direction of the gradient
        in 3 dimensions and so on for more dimensions.
    We repeat this process until we reach the global minimum of the loss function.
    
    Because we are using gradient descent, we have to choose different starting points
        for the weights, because we might get stuck in a local minimum.
