# What is Machine Learning(ML) ?

Machine Learning is having data being fed to the
ML model (mathematical model) and
it giving back some output.
No human is telling the computer what to do.



# Different types of Machine Learnings :

1. Supervised Learning
    When I have data that is already classified (contains the solution)
    We are giving supervision to the model
    Basically we are telling the model this is a 'cat'/'dog'...(supervising)


2. Unsupervised Learning
    We are giving data that is not classified
    
    So we are not saying which data is which(fat, skinny, long, short people)
    but the model is creating/outputting some "clusters"(groups)
    The model is just grouping data that we feed in, we humans have
    the ability to categorize those clusters ourselves


3. Reinforcement
    Works with Agents and Enviornment
    (Basic example human(agent) jumps as high as it can(in the enviornment))
    The agent has actions(human cand bend it's knees, crouch)
    We do not tell the agent what to do, we just reward it for right actions
    Maybe start with random actions(like walking to the left and right the
    agent gets 0 points, if you crouch you recieve -2 points, as soon as the
    person jumps a little +0.5 points, higher jump + 1.5 points)
    Rewards and Punishments so that the model trains itself
    Has a lot to do with genetic algorithms and evolution.
