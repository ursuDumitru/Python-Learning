1. Multithreading

Why use threads ?
You can execute more functions at the same time

```python
import threading

def helloworld():
    print("Hello World!")

t = threading.Thread(target=helloworld)
t.start()
```

Without threads, the function1 is executed first, then function2

if you want the functions to be execute at the same time use
.start() not .run()

# not sure threading works in jupyter notebook
# just copy the cell code inside app_threading.py to see result

```python
def function1():
    for x in range(1000):
        print('ONE')

def function2():
    for x in range(1000):
        print('TWO')

t1 = threading.Thread(target=function1)
t2 = threading.Thread(target=function2)

t1.start()
t2.start()
```

There is a 'main' thread, and every thread we start comes from 'main'

We can specify a thread to wait for another one to finish it's execution

```python
def hello():
    for x in range(50):
        print('Hello World!')

t3 = threading.Thread(target=hello)
t3.start()

t3.join() # don't go to the next thread until my job is finished

print('Just another text') # without any waiting this is executed first
```
